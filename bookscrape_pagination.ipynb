```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url = "https://books.toscrape.com/"

cart = []

for page in range(1, 51):  # Assuming 50 pages for demonstration
    url = f"{base_url}/catalogue/page-{page}.html"  # Modify URL pattern if necessary
    response = requests.get(url)
    response.raise_for_status()

   
    # Parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")
    ol = soup.find("ol")
    articles = ol.find_all("article", class_="product_pod")
    
    
    # Find books listings on the page
    for article in articles:
        title = article.find("img").attrs["alt"]
        rating = article.find("p")
        rating = rating["class"][1]
        price = article.find("p",class_="price_color")
        stock = article.find("p",class_="instock availability")
        url_link = article.find("a")
        url_link = url_link["href"]
    

        cart_title = title if title else "none_t"
        cart_rating = rating if rating else "none_r"
        cart_price = price.get_text() if price else "none_p"
        cart_stock = stock.get_text(strip=True) if stock else "none_s"
        cart_url_link = url_link if url_link else "none_url"


        cart.append(
            {
            "Titile": cart_title,
            "Price": cart_price,
            "Rating": cart_rating,
            "Stock": cart_stock,
            "URL": f"https://books.toscrape.com/catalogue/{cart_url_link}"
            }  
            )


    print(f"Scraped page {page}")  # Indicate progress
```

```python
for display in cart:
    print(display)
```

```python
import pandas as pd
df = pd.DataFrame(cart)
df.index=df.index + 1
# Print the number of movies scraped
print(f"Total books scraped: {len(cart)}")
df
```
